## Script to plot AHN with 100 m resolution (part 2)


```python
# Importing the libraries
import netCDF4 as nc
import matplotlib.pyplot as plt
```

```python
grid = nc.Dataset('mv100.nc')
# Plot using matplotlib.pcolormesh
xx = grid.variables['x'][1000:1500]
yy = grid.variables['y'][500:1000]
zz = grid.variables['depth'][1000:1500, 500:1000]
```

```python
# Plot using imshow
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
plt.imshow(zz[:].T)
plt.colorbar(label='NAP Height [m]')
ax.axis('equal')
ax.axis('tight')
plt.title('Digital Elevation Model for the Nederlands')
plt.xlabel('X-RD [km]')
plt.ylabel('Y-RD [km]')
plt.clim(-50,40)
```

```python
# Plot using matplotlib.pcolormesh
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
cmap = plt.get_cmap('terrain')
plt.pcolormesh(xx[:] / 1000, yy[:] / 1000, zz[:].T, shading='flat',cmap=cmap)
plt.colorbar(label='NAP Height [m]')
ax.axis('equal')
ax.axis('tight')
plt.title('Digital Elevation Model for the Nederlands')
plt.xlabel('X-RD [km]')
plt.ylabel('Y-RD [km]')
plt.clim(-50,40)
```

```python
# Plot using matplotlib.pcolormesh
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
cmap = plt.get_cmap('terrain')
plt.pcolormesh(xx[:] / 1000, yy[:] / 1000, zz[:].T, shading='gouraud',cmap=cmap)
plt.colorbar(label='NAP Height [m]')
ax.axis('equal')
ax.axis('tight')
plt.title('Digital Elevation Model for the Nederlands')
plt.xlabel('X-RD [km]')
plt.ylabel('Y-RD [km]')
plt.clim(-50,40)
```

```python
# 3D plot (better visualised in an IDE rather than Notebook)
%matplotlib notebook
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax = fig.gca(projection='3d')
cset = ax.contourf(xx[:]/1000, yy[:]/1000, zz[:].T/10, cmap='terrain')
plt.show()
```
